```python
l_sum_t = []
l_sum = []
l_sum_t_d = {}

l_log_t = []
l_log = []
l_log_t_d = {}

l_sqrt_t = []
l_sqrt = []
l_sqrt_t_d = {}

i = 1
for l in L:
    # print(l)
    # print('Sum: {}, Mean: {}, norm: {} '.format(np.sum(l), np.mean(l), (np.mean(l) - np.min(l))/ (np.max(l) - np.min(l))))
    l_sum += [(np.mean(l) - np.min(l))/ (np.max(l) - np.min(l))]
    l_sum_t_d[i] = {'len': len(l), 'n': (np.mean(l) - np.min(l))/ (np.max(l) - np.min(l))}
    

    llog = np.log(l)
    # print('Log: {}, Mean: {}, norm: {} '.format(llog, np.mean(llog), (np.mean(llog) - np.min(llog))/ (np.max(llog) - np.min(llog))))
    v_ = (np.mean(llog) - np.min(llog))/ (np.max(llog) - np.min(llog))
    l_log += [v_]
    l_log_t_d[i] = {'len': len(l), 'n': v_ }
    
    
    lsqrt = np.sqrt(l)
    # print('Sqrt: {}, Mean: {}, norm: {} '.format(lsqrt, np.mean(lsqrt), (np.mean(lsqrt) - np.min(lsqrt))/ (np.max(lsqrt) - np.min(lsqrt))))
    v_ =  (np.mean(lsqrt) - np.min(lsqrt))/ (np.max(lsqrt) - np.min(lsqrt))
    l_sqrt +=  [v_]
    l_sqrt_t_d[i] = {'len': len(l), 'n': v_ }
    

    i += 1
    # print('----')
    
l_sum_t = [l_sum]
l_log_t = [l_log]
l_sqrt_t = [l_sqrt]


print(L)
print(l_sum_t)
print(l_log_t)
print(l_sqrt_t)

```

```python
def plot_T2(N):
    # N = 30
    n = np.arange(1, 10)  # n ranges from 1 to 10
    T_values = np.arange(100, 1000, 10) # [0, 100, 1]  # Multiple values for T
    
    # T_values = [300, 292]

    fig, ax = plt.subplots()
    for T in T_values:
        t =  T / (n)
        # Apply min-max normalization
        normalized_t = (t - np.min(t)) / (np.max(t) - np.min(t))
        ax.plot(n, normalized_t, 'o-', label=f'T = {T}')

    ax.set_xlabel('n')
    ax.set_ylabel('t')
    ax.set_title('Plot: t = T/(N*n)')
    ax.grid(True)
    # ax.legend()
    plt.show()


plot_T2(N=1)
```

```python
def cal_t(Tlist):

    x = []
    y = []
    xy = {}
    for T in Tlist:
        n = len(T)
        
        t = np.sum(T) / n
        t = (t - np.min(T)) / (np.max(T) - np.min(T))
        xy[n] = t


        
        x += [n]
        y += [t]
    
    print(xy)
    sxy = {k: xy[k] for k in sorted(xy)}
    print(sxy)

    fig, ax = plt.subplots()
    ax.plot(sxy.keys(), sxy.values(), 'o-', label=f'T = {T}')
    ax.set_xlabel('n')
    ax.set_ylabel('t')
    ax.set_title('Plot: t = T/(N*n)')
    ax.grid(True)
    # ax.legend()
    plt.show()



T_list1 = [
    [300, 250],
    [300, 250, 1000],
    [300, 250, 1000, 3000]
]
cal_t(l_sum_t)
cal_t(l_log_t)
cal_t(l_sqrt_t)
```

```python
def jplot(l_sum_t_d, legd):
    x = [l_sum_t_d[key]['len'] for key in l_sum_t_d]
    y = [l_sum_t_d[key]['n'] for key in l_sum_t_d]

    fig, ax = plt.subplots()
    ax.plot(x, y, 'o-')
    ax.set_xlabel('n')
    ax.set_ylabel('t')
    ax.set_title('Plot: {}'.format(legd))
    ax.grid(True)
    # ax.legend()
    plt.show()

    
jplot(l_sum_t_d, 'sum')
jplot(l_log_t_d, 'log')
jplot(l_sqrt_t_d, 'sqrt')
```
